```python
import pandas as pd
import numpy as np
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
company2013 = pd.read_csv('./data/사업체현황(2017-2019).csv')
```

```python
company2013.head(15)
```

```python
company2013 = company2013.where((company2013['동별(3)'] == '소계') & (company2013['조직형태별(1)'] == '합계') & (company2013['조직형태별(3)'] == '사업체수')).dropna()
```

```python
company2013.drop(company2013.index[0], inplace = True)
```

```python
company2013 = company2013[['동별(2)', '2017', '2018', '2019']]
```

```python
company2013.reset_index(inplace=True)
```

```python
company2013.drop('index', axis = 1, inplace = True)
```

```python
company2013.rename(columns = {'동별(2)':'구'}, inplace = True)
```

```python
company2013['2017'] = company2013['2017'].astype(int)
company2013['2018'] = company2013['2018'].astype(int)
company2013['2019'] = company2013['2019'].astype(int)
```

```python
company2013
```

```python
company2013.to_csv('./data/사업체수2019.csv')
```

```python
df1 = pd.read_csv('./data/김승규.csv')
df2 = pd.read_csv('./data/정현진.csv')
df3 = pd.read_csv('./data/이선희.csv')
df4 = pd.read_csv('./data/김경민.csv')
```

```python
df = pd.concat([df1,df2,df3,df4],axis=0)
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df.index=range(0,4000)
df
```

```python
df['주소']=df['주소'].str.replace('주소','')
df['구']=df['구'].str.replace('서울시','')
df['음식종류']=df['음식종류'].str.replace('음식 종류','')
df['가격대']=df['가격대'].str.replace('가격대','')
```

```python
df
```

```python
wrong = df['음식종류'].str.contains("가격대")
wrong_df = df[wrong]
index = wrong_df.index.to_list()
# print(index)
a = index
a
```

```python
for i in a:
    a = df.loc[i, '음식종류']
    
    df.loc[i, '가격대'] = a
    df.loc[i, '음식종류'] = 0
```
